# 01 - Data Collection

This notebook handles the collection of professional League of Legends player data from Leaguepedia.

## Data Sources
- **Primary**: Leaguepedia (lol.fandom.com) - Uses MediaWiki Cargo API
- **Secondary**: Liquipedia (for supplementary data)

## Tables of Interest
- `Players` - Player biographical info
- `Tenures` - Team history for each player
- `Tournaments` - Tournament/competition information
- `TournamentRosters` - Players participating in tournaments
